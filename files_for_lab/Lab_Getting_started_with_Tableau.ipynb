```python
import pandas as pd
import matplotlib.pyplot as plt
```

### 1.Load the dataset into a notebook.

```python
data= pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
data.head()
```

### 2.Create a barplot of the number of customers per Gender.

```python
#first we should see how many customers we have for each gender

gender_counts = data['Gender'].value_counts()
gender_counts
```

```python

plt.bar(gender_counts.index, gender_counts.values)

plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.title('Number of Customers per Gender')


plt.show()
```

```python
#another method:

import seaborn as sns

sns.countplot(x='Gender', data=data)

plt.title('Number of Customers per Gender')
plt.show()
```

### 3. Create a barplot of the number of customers per EmploymentStatus and Gender.

```python
gender_employment_counts = data.groupby(['EmploymentStatus', 'Gender']).size().unstack()

gender_employment_counts.plot(kind='bar', stacked=True)

plt.xlabel('Employment Status')
plt.ylabel('Number of Customers')
plt.title('Number of Customers per Employment Status and Gender')


plt.show()
```

```python
#another method:

sns.countplot('EmploymentStatus', hue='Gender', data=data)
plt.title('Number of Customers per Employment Status and Gender')
plt.show()
```

#### If we don't want stacked bars:

```python
gender_employment_counts = data.groupby(['EmploymentStatus', 'Gender']).size().unstack()


gender_employment_counts.plot(kind='bar')

plt.xlabel('Employment Status')
plt.ylabel('Number of Customers')
plt.title('Number of Customers per Employment Status and Gender')

plt.legend(['Male', 'Female'])


plt.show()
```
